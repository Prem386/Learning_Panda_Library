üî∞ Step 1: Setting Up and Importing Libraries


```python
# Step 1: Import the essential libraries

import numpy as np           # NumPy for numerical operations
import pandas as pd          # Pandas for data manipulation and analysis
from sklearn.datasets import load_iris  # To load the Iris dataset

# Step 2: Load the Iris dataset using scikit-learn
iris = load_iris()

# Step 3: Convert the dataset to a Pandas DataFrame
# 'data' contains the features; 'feature_names' are the column headers
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Step 4: Add the target (labels) to the DataFrame
# 'target' represents the class/category of each flower
df['species'] = iris.target

# Step 5: Map the numeric target values to actual species names
df['species'] = df['species'].map({i: name for i, name in enumerate(iris.target_names)})

# Step 6: Display the first 5 rows of the dataset
df.head()
```

üîç Step 2: Exploring the Dataset Structure

```python
# View the shape of the dataset
print("Shape of the dataset:", df.shape) #shows dimensions (rows and columns)

# View the column names
print("\nColumn names:\n", df.columns) #list all the features/label names
print()

# View basic information about the dataset
df.info() #prints total entries no of rows, datatypes and null/missing values
print()

# View basic statistical summary (for numerical columns)
df.describe()
print()


#Checking missing values
# Check if any missing values are present in the dataset
df.isnull().sum()




#Viewing values counts(Label Distribution)
# See how many samples there are of each flower species
df['species'].value_counts()



#Viewing Data Types and Conversions
# Check the data types of each column
print(df.dtypes)

# Example: Convert column names to lowercase for uniformity
df.columns = [col.lower().replace(" (cm)", "").replace(" ", "_") for col in df.columns]

# Show updated column names
print("\nUpdated column names:", df.columns)
```

üî∞ Step 3: NumPy Basics ‚Äì Arrays, Shapes, and Operations

```python
#Section A: Importing and Creating Arrays
import numpy as np  # NumPy is usually imported as 'np'

# Create a simple 1D array (vector)
a = np.array([1, 2, 3, 4])
print("1D Array:", a)

# Create a 2D array (matrix)
b = np.array([[1, 2], [3, 4]])
print("\n2D Array:\n", b)

# Create an array of zeros
zeros = np.zeros((2, 3))
print("\nZeros array (2x3):\n", zeros)

# Create an array of ones
ones = np.ones((3, 2))
print("\nOnes array (3x2):\n", ones)

# Create a range of numbers
range_array = np.arange(0, 10, 2)  # start, stop, step
print("\nRange array from 0 to 10 with step 2:", range_array)




#Section B: Array Shape, Size, and Type
# Get shape (rows, columns)
print("\nShape of b:", b.shape)

# Get total number of elements
print("\nSize of b:", b.size)

# Get data type of elements
print("\nData type of a:", a.dtype)




#Section C: Indexing, Slicing, and Reshaping
# Indexing a 1D array
print("First element of a:", a[0])

# Indexing a 2D array
print("Element at 2nd row, 1st column in b:", b[1][0])

# Slicing a 1D array
print("First 3 elements of a:", a[:3])

# Reshaping a 1D array into 2D
reshaped = a.reshape((2, 2))
print("\nReshaped array (2x2):\n", reshaped)




#Section D: Basic Math Operations (Vectorized)
# Element-wise addition
print("\na + 10:", a + 10)

# Element-wise multiplication
print("\na * 2:", a * 2)

# Element-wise power
print("\na squared:", a ** 2)

# Matrix multiplication
c = np.array([[1, 2], [3, 4]])
d = np.array([[5, 6], [7, 8]])
matmul = np.dot(c, d)
print("\nMatrix multiplication:\n", matmul)
```

üî∞ Project using Iris dataset to demonstrate the use of Panda library.

```python
# Importing necessary libraries
import pandas as pd  # for data manipulation
from sklearn.datasets import load_iris  # to load the iris dataset
import matplotlib.pyplot as plt  # for data visualization
import seaborn as sns  # for advanced visualizations

# Load the inbuilt Iris dataset from sklearn
iris = load_iris()

# Convert the iris dataset into a pandas DataFrame
# iris.data contains the features and iris.target contains the class labels
# data is the information in the dataset and feature_name means names of all the columns that are included in the dataset
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add a new column "species" using target names to make the dataset more readable
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
# Categorical.form_codes relaces the iris.target output into iris.target_names ((0,1,3)-->(setosa, vesicolor, virginica))
# in short it maps the numbers to names using a categorical type

# -------------------- BASIC EXPLORATION --------------------

# Show the first 5 rows of the DataFrame
print("üîπ First 5 rows of the dataset:")
print(df.head()) #use tail to display the last 5 rows of the dataset

# Print the shape of the dataset (rows, columns)
print("\nüîπ Dataset shape (rows, columns):", df.shape)

# Check for null/missing values
print("\nüîπ Any missing values?")
print(df.isnull().sum())
# isnull checks whether there are any null values in the dataset. Later we can perform various actions on it if needed
# df.fillna(0, inplace=True)  # fills all nulls with 0

# df.fillna(method='ffill', inplace=True) # fill with fordward fill (ffill): fill nulls with previous non-null values in column

# df.fillna(method='bfill', inplace=True) # fill with backward fill (bfill): fill nulls with next non-null values in column

# df['column_name'].fillna(df['column_name'].mean(), inplace=True) # fill with mean, median or mode for numerical columns

# values = {
#     'column1': 0,
#     'column2': df['column2'].mean(),
#     'column3': 'unknown'
# }
# df.fillna(value=values, inplace=True) # fills different columns with different values

# df.interpolate(method='linear', inplace=True) # useful for time series or numerical data



# Show basic statistical summary of numeric columns
print("\nüîπ Statistical summary:")
print(df.describe())

# -------------------- DATA ANALYSIS --------------------

# Count how many examples of each species are in the dataset
print("\nüîπ Class distribution:")
print(df['species'].value_counts())

# Group by species and calculate mean of each feature
print("\nüîπ Average features per species:")
print(df.groupby('species').mean())

# -------------------- FILTERING AND SORTING --------------------

# Filter rows where sepal length is greater than 6.5
print("\nüîπ Sepal length > 6.5:")
print(df[df['sepal length (cm)'] > 6.5])

# Sort dataset by petal width in descending order
print("\nüîπ Top 5 by petal width:")
print(df.sort_values(by='petal width (cm)', ascending=False).head())

# -------------------- CORRELATION --------------------

# Compute correlation between numerical features
print("\nüîπ Correlation matrix:")
print(df.corr(numeric_only=True))  # numeric_only=True avoids issues with string columns

# -------------------- DATA VISUALIZATION --------------------

# Optional but helpful for visual understanding

# Set Seaborn style
sns.set(style="whitegrid")

# Histogram of petal length for each species
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='petal length (cm)', hue='species', kde=True)
plt.title("Petal Length Distribution by Species")
plt.show()

# Pairplot to see relationships between features
sns.pairplot(df, hue="species")
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()

# Boxplot to compare feature distributions
plt.figure(figsize=(10,6))
sns.boxplot(data=df, orient="h")
plt.title("Boxplots of All Features")
plt.show()

# -------------------- BONUS CHALLENGES --------------------
# Try these if you want to go further:
# 1. Create a new column "petal area" = petal length * petal width
# 2. Save the modified DataFrame to a CSV file
# 3. Create a bar plot showing average sepal length per species

# Example for Bonus 1
df["petal area (cm^2)"] = df["petal length (cm)"] * df["petal width (cm)"]
print("\nüîπ Added 'petal area (cm^2)' column:")
print(df.head())

# Example for Bonus 2
df.to_csv("iris_modified.csv", index=False)
print("\n‚úÖ DataFrame saved to 'iris_modified.csv'.")

# Example for Bonus 3
plt.figure(figsize=(6,4))
sns.barplot(data=df, x='species', y='sepal length (cm)')
plt.title("Average Sepal Length per Species")
plt.show()
```
